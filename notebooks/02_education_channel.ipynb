# 02 · Education channel: persistence → contemporary education

Does institutional persistence predict modern human capital? Regress 2023
educational attainment on the persistence index. A positive coefficient is the
first leg of the human-capital mechanism. **Manuscript: Table 3, column (a).**

```python
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf
from sklearn.preprocessing import StandardScaler

df = pd.read_csv("../data/cleaned_data_v3.csv")
df = df.rename(columns={
    "pct_highschool_or_more (1990)": "pct_hs_1990",
    "pct_highschool_or_more_(2023)": "education_2023",
    "income_1989_real_2023": "income_1989_real",
    "Pop 2010": "pop_2010",
})
df["State"] = df["State"].astype(str).str.strip()

cols = ["Index_v2", "income_1989_real", "pct_hs_1990", "pop_2010", "education_2023"]
df[cols] = df[cols].apply(pd.to_numeric, errors="coerce")
df = df.dropna(subset=cols + ["State"])

df["log_pop_2010"] = np.log(df["pop_2010"])
df[["income_1989_scaled", "pct_hs_1990_scaled"]] = StandardScaler().fit_transform(
    df[["income_1989_real", "pct_hs_1990"]])

model = smf.ols(
    "education_2023 ~ Index_v2 + income_1989_scaled + pct_hs_1990_scaled "
    "+ log_pop_2010 + C(State)",
    data=df).fit(cov_type="HC3")
print(model.summary())
```
