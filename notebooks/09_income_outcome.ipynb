# 09 · Income outcome

Regress log median income (2023) on persistence, with the usual controls.
**Manuscript: §4.**

```python
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf
from sklearn.preprocessing import StandardScaler

df = pd.read_csv("../data/cleaned_data_v3.csv")
df = df.rename(columns={
    "pct_highschool_or_more (1990)": "pct_hs_1990",
    "Pop 2010": "pop_2010",
})

df["log_income_2023"] = np.log(df["income_2023"])
df["log_pop_2010"] = np.log(df["pop_2010"])
df[["income_1989_scaled", "pct_hs_1990_scaled"]] = StandardScaler().fit_transform(
    df[["income_1989_real_2023", "pct_hs_1990"]])

model = smf.ols(
    "log_income_2023 ~ Index_v2 + income_1989_scaled + pct_hs_1990_scaled "
    "+ log_pop_2010 + C(State)",
    data=df).fit(cov_type="HC3")
print(model.summary())
```
